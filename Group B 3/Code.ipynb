```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv("airquality.csv")
data
```

```python
data=data.drop("Unnamed: 0",axis=1)
data
```

```python
data.isnull().sum()
```

```python
data["Ozone"].fillna(data["Ozone"].mean(),inplace=True)
data["Solar.R"].fillna(data["Solar.R"].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
plt.pie([data["Ozone"].mean(),data["Solar.R"].mean(),data["Wind"].mean(),data["Temp"].mean()],
        autopct="%1.0f%%",explode=[0.02,0.02,0.02,0.02],labels=["Ozone","Solar.R","Wind","Temp"])
plt.plot()
```

```python
plt.scatter(data["Temp"],data["Day"],color="red")
```

```python
plt.plot(data.iloc[0:25, 5], data.iloc[0:25, 3], color="purple")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Temperature VS Day")
```

```python
y = np.arange(len(data.iloc[0:25, 3]))

for i in range(len(data.iloc[0:25,3])):
    plt.text(i,data.iloc[0:25,3][i],data.iloc[0:25,3][i],ha='center')

plt.bar(x=y, height=data.iloc[0:25, 3], tick_label=y,width=0.5, color="dodgerblue")
plt.title("Temeprature vs Days Graph")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.boxplot(x=data[["Ozone","Solar.R","Wind","Temp"]],labels=["Ozone","Solar.R","Wind","Temp"])
```

```python
plt.hist(x=data["Temp"],bins=20)
plt.title("Temeprature")
plt.xlabel("Level")
plt.ylabel("Temperature")
```
